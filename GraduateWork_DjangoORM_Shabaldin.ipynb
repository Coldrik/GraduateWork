  **7.1. Сравнение моделей.**

  **Когда Django ORM предпочтительнее:**

  Если вам не нужно сложных запросов: Если ваше приложение не требует сложных операций с фильтрацией или агрегацией, Django ORM будет гораздо удобнее, потому что его фильтры просты и интуитивны.

  Для стандартных CRUD операций: В большинстве случаев Django ORM справляется с задачами фильтрации, создания и обновления записей без необходимости глубоко вникать в SQL.

  Когда нужно быстрое решение: Для быстрого прототипирования или разработки с минимальными усилиями Django ORM будет лучшим выбором.

  **Когда SQLAlchemy может быть полезен:**

  Для сложных запросов и полной гибкости: Если вы работаете с очень сложными базами данных и вам нужно больше контроля над запросами (например, подзапросы, сложные агрегации), SQLAlchemy будет лучше подходить.

  Когда нужно строить динамичные запросы: Если ваше приложение генерирует запросы динамически в зависимости от пользовательского ввода, SQLAlchemy предоставляет инструменты для этого.

  Когда нужно использовать различные базы данных: SQLAlchemy позволяет работать с различными базами данных с минимальными усилиями, что полезно в сложных или многобазовых приложениях.

**Примеры использования преимуществ SQLAlchemy**

*Работа с несколькими базами:*

```python
engine1 = create_engine('sqlite:///db1.sqlite3')
engine2 = create_engine('postgresql://user:password@localhost/db2')

Session1 = sessionmaker(bind=engine1)
Session2 = sessionmaker(bind=engine2)

session1 = Session1()
session2 = Session2()

# Пример объединения данных из двух баз
users = session1.query(User).all()
orders = session2.query(Order).filter(Order.total > 100).all()

# Комбинируем данные
for user in users:
    user_orders = [order for order in orders if order.user_id == user.id]
    print(user.username, len(user_orders))
```

*Динамическое объединение нескольких таблиц:*

В отличии от SQLAlchemy, Django ORM требует заранее определенных связей через ForeignKey или related_name, и объединения должны быть явно указаны. SQLAlchemy позволяет динамически указывать, какие таблицы объединять.

```python
from sqlalchemy.orm import aliased
from sqlalchemy import or_

# Допустим, у нас есть модели User и Order, связанные через ForeignKey

def get_users_with_orders(filter_field=None, filter_value=None):
    OrderAlias = aliased(Order)  # Создаем псевдоним для таблицы Order

    query = (
        session.query(User, OrderAlias)  # Объединяем User с псевдонимом Order
        .join(OrderAlias, User.id == OrderAlias.user_id)  # Делаем JOIN
    )

    if filter_field and filter_value:
        # Применяем фильтр к любой колонке
        if filter_field == 'username':
            query = query.filter(User.username.ilike(f"%{filter_value}%"))
        elif filter_field == 'order_name':
            query = query.filter(OrderAlias.name.ilike(f"%{filter_value}%"))

    return query.all()

# Пример использования
users_with_orders = get_users_with_orders('order_name', 'book')
for user, order in users_with_orders:
    print(user.username, order.name)
```

**7.2.3. Описание преимуществ и недостатков ORM библиотек**

**Django ORM**

Преимущества:

*   Интеграция с Django: Поскольку Django ORM является частью фреймворка Django, он идеально интегрирован в экосистему, что упрощает разработку веб-приложений.
*   Удобство использования: Обладает интуитивно понятным API, что делает его доступным для новичков.
*   Автоматическое создание админки: Позволяет быстро создать интерфейс для управления данными без дополнительной настройки.
*   Миграции: Поддерживает автоматическое управление схемой базы данных с помощью миграций, что упрощает обновления и изменения структуры данных.
*   Сильная документация: Обширная и хорошо структурированная документация, что упрощает процесс изучения.

Недостатки:

*   Производительность: В некоторых случаях может быть менее эффективным по сравнению с другими ORM, особенно при выполнении сложных запросов.
*   Ограниченная гибкость: Более строгий подход к архитектуре может затруднить выполнение нестандартных запросов.
*   Зависимость от Django: Использование Django ORM ограничивает вас экосистемой Django, что может быть проблемой при использовании других фреймворков.

**SQLAlchemy**

Преимущества:

*   SQLAlchemy следует выбирать при работе с несколькими базами данных в рамках одного приложения
*   Возможно реализовать индивидуальный подход к миграции
*   Гибкость: Предоставляет более низкоуровневый доступ к SQL, позволяя разработчикам создавать сложные запросы и управлять базой данных более точно.
*   Разнообразие стилей: Поддерживает как декларативный, так и императивный стиль, что даёт возможность выбрать удобный для вас подход.
*   Отличная производительность: Благодаря оптимизации запросов и кэшированию может быть более производительным в сложных сценариях.
*   Поддержка асинхронности: С недавних пор поддерживает асинхронные операции, что важно для современных приложений.
*  

Недостатки:

*   Сложность: Имеет более крутой порог входа по сравнению с другими ORM, что может затруднить обучение новичкам.
*   Необходимость ручной настройки: В некоторых случаях может потребоваться больше усилий для настройки и конфигурирования.
*   Менее структурированный: Гибкость иногда приводит к менее структурированному коду, если разработчики не следуют лучшим практикам.
*   Для данного проекта пришлось бы отдельно писать блок кода для регистрации и аутентификации пользователей
*   В рамках реализации данного проекта пришлось бы много писать вручную, например, хеширование паролей, проверки на существование пользователя, управление сессиями и т. д.

**Tortoise ORM**

Преимущества:

*   Асинхронная поддержка: Разработан с учётом асинхронного программирования, что делает его отличным выбором для современных приложений на Python.
*   Простота использования: API очень похож на Django ORM, что облегчает переход для разработчиков, знакомых с Django.
*   Современные возможности: Поддерживает аннотации типов, что улучшает разработку и читаемость кода.
*   Лёгкая интеграция: Легко интегрируется с различными фреймворками, такими как FastAPI и Sanic.

Недостатки:

*   Меньше функций: В сравнении с более зрелыми ORM, такими как Django ORM и SQLAlchemy, может недоставать некоторых функций и возможностей.
*   Меньшая экосистема: Имеет менее развитое сообщество и документацию по сравнению с Django ORM и SQLAlchemy.
*   Проблемы с производительностью: В некоторых случаях может показывать худшие результаты по производительности в зависимости от используемого подхода.

Если ваше приложение ориентировано на асинхронность и вы хотите использовать асинхронные возможности Django, Tortoise ORM станет отличным выбором, потому что он нативно поддерживает асинхронные запросы и операции с базой данных, в отличие от Django ORM, который является синхронным.

Если вам важна поддержка синхронного кода и вы работаете в традиционной модели Django, то лучше использовать Django ORM. Но для современных приложений, работающих с асинхронными задачами и большими объемами данных, Tortoise ORM может быть более удобным и эффективным выбором.
